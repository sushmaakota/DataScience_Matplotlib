## Exercise 1: Read Total profit of all months and show it using a line plot

* X label name = Month Number
* Y label name = Total profit

</div>

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

dataset = pd.read_csv("company_sales_data.csv")
dataset.head()
```

```python
# Exercise-1 code here
profit = dataset ['total_profit'].tolist()
month = dataset ['month_number'].tolist()
plt.plot(month, profit, label = 'Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(month)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

```python
# Exercise-8 code here
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [dataset ['facecream'].sum(),dataset ['facewash'].sum(), dataset ['toothpaste'].sum(), 
         dataset ['bathingsoap'].sum(), dataset['shampoo'].sum(), dataset['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```
